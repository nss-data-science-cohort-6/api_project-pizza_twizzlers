```python
import requests
import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
#Question 2: Census
```

```python
endpoint = 'https://api.census.gov/data/2020/acs/acs5/variables.html'

response = requests.get(endpoint)
```

```python
tn_census = f'https://api.census.gov/data/2020/acs/acs5?get=NAME,B01001_001E&for=tract:*&in=state:47'

response = requests.get(tn_census)

response.text
```

```python
tn_census_json = response.json()
tn_census_json
```

```python
tn_davidson_census = pd.DataFrame(response.json())
tn_davidson_census
```

```python
tn_davidson_census_df = tn_davidson_census.rename(columns ={0:'Name', 1: 'Census', 2: 'State_Code', 3: 'County_Code', 4: 'Tract'})
```

```python
tn_davidson_census_df= tn_davidson_census_df.drop(labels=0, axis=0)
tn_davidson_census_df
```

```python
tn_davidson_census_df_TS = tn_davidson_census_df[tn_davidson_census_df['County_Code'] == '037']
tn_davidson_census_df_TS
```
